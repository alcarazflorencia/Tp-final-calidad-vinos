**Outliers**

```python
def detect_outliers(df):
    outlier_indices = []
    for col in df:
        if df[col].dtype == np.float64 or df[col].dtype == np.int64:
            Q1 = np.percentile(df[col], 25)
            Q3 = np.percentile(df[col], 75)
            IQR = Q3 - Q1
            outlier_step = 1.5 * IQR
            outlier_list_col = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step)].index
            outlier_indices.extend(outlier_list_col)
    return outlier_indices
outlier_indices = detect_outliers(vinos_df)
outlier_data = vinos_df.iloc[outlier_indices]
outlier_data.head()

##seleccion
```

```python
Arriba podemos identificar los vinos que contienen por lo menos, algina caracteristica que se denomine como atipica/valor atipico; acá podemos identificar por el iqr 
```

```python

columnas_numericas = vinos_df.select_dtypes(include=['float64', 'int64']).columns

z_scores = stats.zscore(vinos_df[columnas_numericas])
umbral_outliers = 3
outliers = np.where(np.abs(z_scores) > umbral_outliers)
print(f'Cantidad de outliers en el conjunto de datos: {len(outliers[0])}')
```

```python
## Seguimos con el grafico

sns.boxplot(data=outlier_data,palette='pastel')
plt.title('Valor atipico significativo')
plt.xticks(rotation=45)
plt.show()
```

```python
Vemos los valores atipicos segun el conjunto de datos, en este caso el total de dioxido de azufre es la variable que mas valores atipicos 
contiene que se aleja en gran proporcion de la mayoria.
```

**Regresionn**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Variable#### revisar
X_simple = vinos_df[['alcohol']]
y_simple = vinos_df['quality']

X_train_simple, X_test_simple, y_train_simple, y_test_simple = train_test_split(X_simple, y_simple, test_size=0.2, random_state=42)
model_simple = LinearRegression()
model_simple.fit(X_train_simple, y_train_simple)
predictions_simple = model_simple.predict(X_test_simple)
print('Coeficientes: \n', model_simple.coef_)
mse = mean_squared_error(y_test_simple, predictions_simple)
print('Error cuadratico:', mse)

#
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_test_simple['alcohol'], y_test_simple, predictions_simple, c='m', marker='+', label='Datos')

ax.set_xlabel('Alcohol')
ax.set_ylabel('Calidad')
ax.set_zlabel('Prediccion calidad')
plt.show()
```

```python
En este grafico,  el eje horizontal x representa la cantidad de alcohol en el vino, el eje vertical y, muestra la calidad del vino, y el eje que se adentra en el gráfico representa la calidad predicha.
Los puntos en el gráfico representa una muestra de vino, se observa que a medida que aumenta el contenido de alcohol, también tiende a aumentar la calidad de l vino
```

**Conclusion**


El análisis de regresión indico que el contenido de alcohol podría ser un indicador significativo de la calidad del vino.

la calidad del vino tiende a hacerlo también.
